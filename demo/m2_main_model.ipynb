```python
from tqdm import tqdm

from IPython.display import clear_output
%matplotlib inline
%config InlineBackend.figure_format='retina'

import sys
sys.path.append('..')
from utils.env_fn import *
from utils.model import *
from utils.fig_fn import *
```

```python
def sim_choice(board, player_to_move, n_sim, agent=heuristic_agent):
    p_actions = {}
    for i in tqdm(range(n_sim)):
        model = agent(env=four_in_a_row(), 
                            params=default_params().to_list())
        action = model.get_action((board, player_to_move))
        p_actions[action] = p_actions.get(action, 0) + 1

    show_board(board, p_actions)
```

```python
# Condition 1
layout = [
    '.....10..',
    '.....11..',
    '..10001..',
    '.10001...',
]
n_sim, player_to_move = 100, 0
sim_choice(layout2board(layout), player_to_move, n_sim, agent=BFS_agent)
```
